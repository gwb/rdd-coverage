```julia
n = 1000
x = randn(n)
y = randn(n)
k = SEIso(0.0,0.0)
logNoise=log(10)
m = MeanZero()
gp = GP(x', y, m, k, logNoise)
;
```

```julia
KXstar_X = cov(gp.k, [0.0]', gp.X)
Ky = gp.cK
weights = vec(gp.cK \ KXstar_X')
;
```

```julia
plt.plot(x, weights, ".")
```
