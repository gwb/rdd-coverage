```r
#Create raw data
cov.pars.true = c(2.0, 0.2)
sim_residuals <- grf(500, cov.pars = cov.pars.true, cov.model="exponential", nugget=.1)
trend = 0*sim_residuals$coords[,1]
sim_data <- sim_residuals
sim_data$data <- sim_data$data + trend
par(mfrow=c(1,2))
points.geodata(sim_residuals, main = "Simulated locations and values") 
points.geodata(sim_data, main = "After adding trend") 
```

```r
# Let's look at the empirical variograms
par(mfrow = c(2, 2)) 
emp_vario1 <- variog(sim_data, option = "cloud", max.dist=1) 
  plot(emp_vario1, main = "Classical estimator") 
emp_vario2 <- variog(sim_data, option = "cloud", estimator.type = "modulus", max.dist=1)
  plot(emp_vario2, main = "Cressie estimator") 
bin_vario1 <- variog(sim_data, uvec = seq(0, 1, l = 12), bin.cloud = T, max.dist=1) 
  plot(bin_vario1, main = "Classical estimator")
lines.variomodel(cov.model = "exponential", 
    cov.pars = cov.pars.true, nug = 0.1, max.dist=1)
bin_vario2 <- variog(sim_data, uvec = seq(0, 1, l = 12), bin.cloud = T, max.dist=1, estimator.type = "modulus") 
  plot(bin_vario2, main = "Cressie estimator")
lines.variomodel(cov.model = "exponential", 
    cov.pars = cov.pars.true, nug = 0.1, max.dist=1)  
```

```r
# Fit a parametric variogram

# OLS
parametric_vario_ols <- variofit(bin_vario2, ini.cov.pars = c(1, 1), 
    cov.model= "exponential", weights = "equal") 
summary(parametric_vario_ols)
# WLS
parametric_vario_wls <- variofit(bin_vario2, ini.cov.pars = c(1, 1), 
    cov.model= "exponential") 
summary(parametric_vario_wls )
# ML
parametric_vario_ml <- likfit(sim_data, ini.cov.pars = c(1, 1), 
    cov.model= "exponential", trend="cte")
summary(parametric_vario_ml )
```

```r
# Plot the fits vs empirical and true values
par(mfrow=c(1,1))
plot(bin_vario2)
lines.variomodel(cov.model = "exponential", 
    cov.pars = cov.pars.true, nug = 0.1, max.dist=1)
lines(parametric_vario_ols, lwd=2, lty="dotted")
lines(parametric_vario_wls, lwd=2, lty="dashed")
lines(parametric_vario_ml, lwd=2, lty="dotdash")
legend(0.8, 1.1, c("True", "OLS", "WLS", "ML"), lty = c("solid", "dotted","dashed","dotdash"))
```

```r
# Use simulation to test model fit
env.mc <- variog.mc.env(sim_data, obj.var = bin_vario2) 
env.model <- variog.model.env(sim_data, obj.var = bin_vario2, model.pars = parametric_vario_wls)
par(mfrow = c(1, 2)) 
plot(bin_vario1, envelope = env.mc) 
plot(bin_vario1, envelope = env.model)
```

```r
# Plot the resulting surface
#Set up a grid of predicted values
pred.grid <- expand.grid(seq(0,1, l = 51), seq(0,1, l = 51)) 
kc <- krige.conv(sim_data, loc = pred.grid, krige = krige.control(obj.m = parametric_vario_ml))
# 2 different plots
par(mfrow=c(1,2))
image(kc, loc = pred.grid, col = gray(seq(1, 0.1, l = 30)), xlab = "Coord X", ylab = "Coord Y")
points(sim_data, xlab = "Coord X", ylab = "Coord Y", cex.max = 1.7, col = gray(seq(1, 0.1, l = 100)))
contour(kc, loc = pred.grid, xlab = "Coord X", ylab = "Coord Y", add=T)
```
