# Sharp vs Weak test power

```julia
import PyPlot; plt=PyPlot
plt.rc("figure", dpi=300.0)
# plt.rc("figure", figsize=(6,4))
plt.rc("figure", autolayout=true)
plt.rc("savefig", dpi=300.0)
plt.rc("text", usetex=true)
plt.rc("font", family="serif")
plt.rc("font", serif="Palatino")
using LaTeXStrings
;
```

\begin{equation}
X_i \overset{\indep}{\sim} \normal\del{\mu_i, 1}\text{, for }i=1,\ldots,n
\end{equation}
\begin{equation}
\begin{split}
    H_{0,\mathrm{sharp}} &: \mu_i = 0 &\text{ , the sharp (Fisherian) null hypothesis} \\
    H_{0,\mathrm{weak}} &: M = \frac{1}{n} \sum_{j=1}^n\mu_i = 0 &\text{ , the  weak (Neymenian) null hypothesis} \\
\end{split}
\end{equation}
Hypothesis test for the sharp null:
\begin{equation}
\begin{split}
    S &= \sum_{j=1}^n X_i^2 \\
    S \mid H_{0,\mathrm{sharp}} &\sim \chi^2_p
\end{split}
\end{equation}
Hypothesis test for the weak null \\
\begin{equation}
\begin{split}
    t &= \frac{1}{n}\sum_{j=1}^k X_i \\
    t \mid H_{0,\mathrm{weak}} &\sim \normal\del{0, \frac{1}{n}}
\end{split}
\end{equation}

Under the sharp null, both tests set with significance level $\alpha$ will have power $\alpha$. The question is, under a constant effect $\mu_i=\tau$, which test will have greater power? We'll answer the question through simulations, and then theoretically.

## Power Simulations

```julia
using Distributions
```

```julia
function sim_p_meantest(τ::Float64,n::Int)
    Y=randn(n)+τ
    t=mean(Y)
    Dnull = Normal(0,1/√n)
    return 2*cdf(Dnull,-abs(t))
end
```

```julia
function sim_p_chitest(τ::Float64,n::Int)
    Y=randn(n)+τ
    S=sum(Y.^2)
    Dnull = Chisq(n)
    return ccdf(Dnull,S)
end
```

```julia
sim_p_chitest(0.0,10)
```

```julia
τ=0.3
n=10
p_meantest_null = [sim_p_meantest(0.0,n) for _ in 1:100000]
p_meantest_τ0_5 = [sim_p_meantest(τ,n) for _ in 1:100000]
p_chitest_null = [sim_p_chitest(0.0,n) for _ in 1:100000]
p_chitest_τ0_5 = [sim_p_chitest(τ,n) for _ in 1:100000]
;
```

```julia
plt.plt[:hist](p_meantest_null, histtype="step", cumulative=true, bins=linspace(0,1,1000), normed=true)
plt.plt[:hist](p_meantest_τ0_5, histtype="step", cumulative=true, bins=linspace(0,1,1000), normed=true, label="mean")
plt.plt[:hist](p_chitest_null, histtype="step", cumulative=true, bins=linspace(0,1,1000), normed=true)
plt.plt[:hist](p_chitest_τ0_5, histtype="step", cumulative=true, bins=linspace(0,1,1000), normed=true, label=L"\chi^2")
plt.xlim(0,1)
plt.ylim(0,1)
plt.legend(loc="best")
plt.title("The mean test (Neymenian test targetting the weak null) has higher power")
;
```

## Theoretical power

Under the alternative hypothesis of a constant effect $\tau$, $t$ (the **Neymenian** test statistic) has distribution
\begin{equation}
    \del{t \mid \mu_i = \tau} \sim \normal\del{\tau, 1 / p}
\end{equation}
The threshold $t^*$ beyond which the null hypothesis is rejected with significance level $\alpha$ is
\begin{equation}
    t^* = \abs{ \frac{1}{\sqrt{p}} \Phi^{-1}\del{\alpha \over 2}}
\end{equation}
and thus the power is given by the CDF of the normal distribution.

Turning our attention to the Fisher sharp null test, the distribution under the alternative is a non-central $\chi^2$
\begin{equation}
    \del{S \mid \mu_i = \tau} \sim \chi^2_p \del{\tau p}
\end{equation}
and the threshold $S^*$ is the $1-\alpha$ quantile of the $\chi^2_p$ distribution. Power is correspondingly given by the CDF of the non-central $\chi^2$ distribution.

We compute the power of the two test with $n=10$, $\alpha=0.05$ and $\tau=0.3$, and get:

* power of the Neymanian test is 15.78%
* power of the Fisherian test is 7.85%

This is close (within simulation error) of the simulated power values.

```julia
α=0.05
null_distr=Normal(0,1/√n)
thresh=abs(quantile(null_distr,α/2))
altv_distr=Normal(τ,1/√n)
power=cdf(altv_distr,-thresh)+ccdf(altv_distr,thresh)
println("Mean test")
@printf("theoretical power: %.2f%%\n", power*100)
@printf("empirical power: %.2f%%", mean(p_meantest_τ0_5.<α)*100)
```

```julia
α=0.05
null_distr=Chisq(n)
thresh=quantile(null_distr,1-α)
altv_distr=NoncentralChisq(n,n*τ^2)
power=ccdf(altv_distr,thresh)
println("χ² test")
@printf("theoretical power: %.2f%%\n", power*100)
@printf("empirical power: %.2f%%", mean(p_chitest_τ0_5.<α)*100)
```

We therefore observe both through simulations and theoretical values that the power of the Fisher exact test is lower than the Neymanian weak null hypothesis test.

# Consistency of $\gp$s with misspecified covariance parameters

```julia
using GaussianProcesses
```

```julia
n = 10000
x = 10*randn(n)
y = randn(n)
k1 = SEIso(0.0,0.0)
m = MeanZero()
logNoise=1.0
```

```julia
gp1_nonoise = GP(;m=m,k=k1,logNoise=log(1e-8))
```

```julia
f_from_k1 = rand(gp1, x');
y_from_k1 = f_from_k1 + randn(n)
;
```

```julia
k100 = SEIso(log(2.0),0.0)
gp100 = GP(x',y_from_k1,m,k100,0.0);
```

```julia
xx = linspace(-10,10,100)
pred_100 = predict(gp100, xx')
```
